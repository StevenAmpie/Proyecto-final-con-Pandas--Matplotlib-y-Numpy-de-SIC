# Halle el valor máximo de las ausencias y considere dicho valor como el total de clases del curso, de manera que pueda sacar un porcentaje de asistencia para cada estudiante

## Porcentajes de asistencias en el curso de Matematicas para GP; cree una nueva columna llamada “extra”

```python
gp_math['Porcentaje de asistencia'], gp_por['Porcentaje de asistencia'], ms_math['Porcentaje de asistencia'], ms_por['Porcentaje de asistencia']='','','',''  
gp_math['extra'], gp_por['extra'], ms_math['extra'], ms_por['extra']='','','',''  #CREACION DE LA COLUMNA EXTRA EN CADA UNO DE LOS DATAFRAMES
```

```python
gp_math['absences'].max() #aqui se ha hallado el valor maximo de las ausencias, es decir que la persona con mas ausencias, posee 75
```

```python
porcentajes_gp_math=[]
for i in gp_math['absences']:
    por=(75-i)/75*100
    porcentajes_gp_math.append(round(por,2))
```

```python
for i in porcentajes_gp_math:
    print(i, '%', end=', ' )   
```

```python
gp_math['Porcentaje de asistencia']=porcentajes_gp_math
```

# Cree una nueva columna llamada “approved” en la cual determine si el estudiante aprueba o reprueba el curso (1 para aprueba, 0 para reprueba), considerando:
#### •	Si el estudiante tiene un porcentaje de asistencia menor al 80% del curso reprueba D
#### •	Si el estudiante tiene un porcentaje de asistencia mayor al 80% del curso, pero la nota G3 es menor a 10 reprueba    C
#### •	Si el estudiante tiene un porcentaje de asistencia mayor al 80% del curso, pero la nota G3 está entre 10 y 15 aprueba y se asigna 1 a la columna extra B
#### •	Si el estudiante tiene un porcentaje de asistencia mayor al 80% del curso y la nota G3 es mayor a 15 aprueba y se asigna 0 a la columna extra A


```python
from numpy import select
valores1=['','',0,0] #valores para los reprobados
valores2=[0,1,'',''] #valores para los aprobados
```

## gp_math

```python
condiciones_para_reprobados_gp_math=[
    (gp_math['Porcentaje de asistencia']>80) & (gp_math['G3']>15),
    (gp_math['Porcentaje de asistencia']>80)&(gp_math['G3']>=10) & (gp_math['G3']<=15), 
    (gp_math['Porcentaje de asistencia']>80) & (gp_math['G3']<10),
    (gp_math['Porcentaje de asistencia']<80)  
    ]
    
condiciones_para_aprobados_gp_math=[
    (gp_math['Porcentaje de asistencia']>80) & (gp_math['G3']>15),
    (gp_math['Porcentaje de asistencia']>80)&(gp_math['G3']>=10) & (gp_math['G3']<=15), 
    (gp_math['Porcentaje de asistencia']>80) & (gp_math['G3']<10),
    (gp_math['Porcentaje de asistencia']<80)  
    ]
```

```python
gp_math['approved']=select(condiciones_para_reprobados_gp_math, valores1)
gp_math['extra']=select(condiciones_para_aprobados_gp_math, valores2)
```
